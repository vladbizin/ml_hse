# XOR set

## Problem setting
You are given a labeled XOR set.  

## Question
Can you construct a dividing surface with linear regression and polynomial features?

## Solution

It is obvious that the dividing surface is given by eqation xy=0. Excellent, then we use the polynomial feature xy.  
(In the example below i add a fit intercept term just to plot the graphs, the model itself works perfectly without it.)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
```

```python
x = np.random.uniform(-10, 10, size=(2000, 2))
y = (x[:, 0] * x[:, 1] >0).astype('int')

plt.scatter(x[:, 0], x[:, 1], color=['red' if c==1 else 'blue' for c in y], s = 5)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
log_reg = LogisticRegression()
log_reg.fit(np.array([x[:, 0] * x[:, 1]]).T, y,);
```

```python
# axy + b = 0
a = log_reg.coef_[0][0]
b = log_reg.intercept_[0]

surface_x = np.linspace(-10.5, 10.5, 5000)
surface_y = -b/a/surface_x
```

```python
plt.scatter(x[:, 0], x[:, 1], color=['red' if c==1 else 'blue' for c in y], s = 5)
plt.plot(surface_x, surface_y, color='green', linewidth=2)
plt.ylim(-10.5, 10.5)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
